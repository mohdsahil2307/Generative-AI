```python
import pandas as pd
messages = pd.read_csv("datasets/spam.csv", encoding="latin1")
```

```python
messages = messages[['v1','v2']]
```

```python
messages.rename(columns={'v1':'output','v2':'message'},inplace = True)
```

```python
import re
import nltk
nltk.download('stopwords')
```

```python
from nltk.stem import PorterStemmer
from nltk.corpus import stopwords
ps = PorterStemmer()
```

```python
corpus = []
for i in range(0,len(messages)):
    text = messages['message'][i].lower()
    text = re.sub('[^a-zA-Z]',' ',messages['message'][i])
    text = text.split()
    text = [ps.stem(word) for word in text if word not in stopwords.words('english')]
    text = ' '.join(text)
    corpus.append(text)
```

```python
corpus
```

# BAG OF WORDS : 

```python
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(binary=True, max_features=100)
```

```python
X = cv.fit_transform(corpus).toarray()
```

```python
X.shape
```

# NGRAMS

```python
cv.vocabulary_
# THIS PRINTS THE TOP 100 WORDS (DEPENDING UPON THE MAX_FEATURES USED, AND GIVE THEIR INDEXES)
```

```python
#NOW WE IMPLEMENT COUNT VECTORIZER WITH N_GRAMS : 
cv = CountVectorizer(binary=True, max_features=500, ngram_range=(1,2)) # THE NGRAM RANGE TELLS IT TO USE UNI AND BIGRAMS
X= cv.fit_transform(corpus).toarray()
```

```python
cv.vocabulary_ # NOW IT CONTAINS COMBINATION OF TWO WORDS AS WELL
```
